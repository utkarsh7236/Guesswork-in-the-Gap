```python
import pandas as pd
import numpy as np
import gzip
from tqdm import tqdm

def read_csv_gz(file_path):
    with gzip.open(file_path, 'rt') as f:
        data = np.genfromtxt(f, delimiter=',', names=True, dtype=None, encoding='utf-8')
    return data
```
