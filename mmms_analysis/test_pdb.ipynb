```python
EVENT_ARGS = ""
EVENT_ARGS += " --event-max-num-samples 10000 "
EVENT_ARGS += " --prior-column logprior "
EVENT_ARGS += " --prior-is-log "
EVENT_ARGS += " --m1-column mass1_source "
EVENT_ARGS += " --m2-column mass2_source "
EVENT_ARGS += " --d-column luminosity_distance "
EVENT_ARGS += " --q-range 0.0 1.0 "
EVENT_ARGS += " --m-range 0.0 10.0 "
EVENT_ARGS += " --mc-range 0.0 10.0 "
EVENT_ARGS += " --d-range 0.0 10000.0 "
```

```python
POP_ARGS = ""
POP_ARGS += " --pop-max-num-samples 10000 "
POP_ARGS += " --mtov-column Mmax "
POP_ARGS += " --rtov-column Rmax "
POP_ARGS += " --pop-weight-is-log "
```

```python
SEED = " --seed 7236"
```

```python
EOS_WEIGHT = " logweight_PSR_GW_Xray"
EVENT_SAMPLES = "Prod-BBH-HighSpin1-HighSpin2"
POP_SAMPLES = "Farah2022-g-PDB"
EOS_SAMPLES = "LEC-2020"
COMPONENT = "1"

EXTRA_EVENT_ARGS=f"--mass-column mass{COMPONENT}_source --spin-column spin{COMPONENT}_magnitude"
EXTRA_POP_ARGS=f"--pop-weight-column {EOS_WEIGHT}"

LABEL= f"{EVENT_SAMPLES}+{POP_SAMPLES}+{EOS_SAMPLES}-{EOS_WEIGHT}+component{COMPONENT}"

assert os.path.isfile(f"etc/{POP_SAMPLES}+{EOS_SAMPLES}.csv.gz")
assert os.path.isfile(f"etc/{EVENT_SAMPLES}.csv.gz")
```

```python
cmd_str = (f"mmms etc/{EVENT_SAMPLES}.csv.gz "
           f"{POP_SAMPLES}.ini "
           f"etc/{POP_SAMPLES}+{EOS_SAMPLES}.csv.gz"
           f"{EVENT_ARGS}"
           f"{EXTRA_EVENT_ARGS}"
           f"{POP_ARGS}"
           f"{EXTRA_POP_ARGS}"
           f"{SEED}"
           f" --Verbose "
           f"1> testing_mmms/{LABEL}.out "
           f"2> testing_mmms/{LABEL}.err ")

os.system(cmd_str)
```

```python
cmd_str2 = (f" mmms-plot"
            f"etc/{EVENT_SAMPLES}.csv.gz"
            f"EVENT_ARGS"
            f"EXTRA_EVENT_ARGS"
            f"{POP_SAMPLES}.ini"
            f"etc/{POP_SAMPLES}+{EOS_SAMPLES}.csv.gz"
            f"POP_ARGS"
            f"EXTRA_POP_ARGS"
            f"SEED"
            f"--Verbose"
            f"--output-dir figures"
            f"--tag {LABEL}"
            f"--figtype png --dpi 200")

os.command(cmd_str2)
```
