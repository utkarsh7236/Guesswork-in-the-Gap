```python
import os
import pandas as pd
import gzip
import glob
```

```python

target_filename = "population.csv.gz"
root_dir = os.getcwd()  # current directory

missing_dirs = []

for dirpath, dirnames, filenames in os.walk(root_dir):
    # Skip the root itself
    if "__pycache__" in dirpath.split(os.sep):
        continue
    if dirpath == root_dir:
        continue
    if target_filename not in filenames:
        missing_dirs.append(dirpath)

if missing_dirs:
    raise FileNotFoundError(
        f"{target_filename} not found in the following directories:\n" +
        "\n".join(missing_dirs)
    )
else:
    print("[STATUS] All folders contain population.csv.gz")
```

```python
# Define the required columns
required_columns = {"Ncomp", "OmegaMatter", "OmegaRadiation", "OmegaLambda", "max_redshift"}

# Get all population.csv.gz files (recursively, if needed)
files = glob.glob("**/population.csv.gz", recursive=True)

# Assert each file contains all required columns
for file in files:
    with gzip.open(file, 'rt') as f:
        df = pd.read_csv(f, nrows=0)  # Only read header
        missing = required_columns - set(df.columns)
        assert not missing, f"{file} is missing columns: {missing}"

print("All files contain the required columns.")
```
