```python
import os
```

```python

target_filename = "population.csv.gz"
root_dir = os.getcwd()  # current directory

missing_dirs = []

for dirpath, dirnames, filenames in os.walk(root_dir):
    # Skip the root itself
    if "__pycache__" in dirpath.split(os.sep):
        continue
    if dirpath == root_dir:
        continue
    if target_filename not in filenames:
        missing_dirs.append(dirpath)

if missing_dirs:
    raise FileNotFoundError(
        f"{target_filename} not found in the following directories:\n" +
        "\n".join(missing_dirs)
    )
else:
    print("[STATUS] All folders contain population.csv.gz")
```
